# 주성분 분석 PCA

**주성분 분석** : 다변량 자료에 대해 서로 독립적인 주요 성분을 도출하여 원래 변수들이 가지고 있는 변이(정보)를 축약 <br>
<br>
**분산**: 일변량 자료가 가지는 변이의 크기 <br>
**다변량에서 변이**: 공분산, 상관계수행렬

**대표값(대표하는 특성치의 예)** <br>
<br>
산술평균: $$\bar{X} = {1 \over p}\sum{X_{i}}$$
가중평균: $$\bar{X_{a}} = \sum{a_{i}X_{i}}$$
어떠한 함수(일반화): $$\bar{X_{f}} = f(\sum{X_{i}})$$

대표한다는 것은 **총변이를 가장 많이 함축하는 것**

**주성분 분석(PCA: Principal Component Analysis)** 이란 서로 연관되어 있는 변수들이 전체적으로 가지고 잇는 정보들을 최대한 확보하는 작은 수의 새로운 변수(주성분)을 생성해내는 방법이다. <br>
원래 있던 변수들의 선형결합을 통해 주성분을 도출해낸다. <br>
주성분분석에는 주성분을 도출하는 방법과 생성된 주성분의 성격, 이에 연관된 통계적 추론이 포함된다.

**회귀분석에서 설명변수의 수가 많을 때** <br>
변수 선택방법대신 주성분 분석을 실시하여 차원을 축소, 새로운 변수를 만들어주었을 때 더 간단하고 좋은 모형을 만들 수 있다.


만약 10 x 10의 자료가 있을 때
$$\bar{X_{1}} = {1 \over 10}\sum{X_{1i}}$$
$$\bar{X_{2}} = {1 \over 10}\sum{X_{2i}}$$
와 같이 정의하면
$$s_{11} = {1 \over 10-1}\sum{(X_{1i} - \bar{X_{1}})^2}$$
$$s_{22} = {1 \over 10-1}\sum{(X_{2i} - \bar{X_{2}})^2}$$
$$COV(X_{1}, X_{2}) = s_{12} = {1 \over 10-1}\sum{(X_{1i} - \bar{X_{1}})(X_{2i} - \bar{X_{2}})}$$

상관계수는
$$r_{12} = {s_{12} \over \sqrt{s_{11}}\sqrt{s_{22}}}$$

이와 같이 공분산과 분산을 원소로 갖는 분산 공분산 행렬로 나타낼 수 있다.
$$S = \begin{matrix} 
     & X_{1} & X_{2}& X_{3}& X_{4}& X_{5} \\
    X_{1} & s_{11} & s_{12} & s_{13} &s_{14} &s_{15} \\
    X_{2} &  & s_{22} & s_{23} & s_{24} & s_{25} \\
    X_{3} & & & s_{33} & s_{34} & s_{35} \\
    X_{4} & 대 & 칭 & & s_{44} & s_{45} \\
    X_{5} & & & & & s_55
\end{matrix}$$
마찬가지로 상관계수 행렬도 구할 수 있다.
$$R = \begin{matrix} 
     & X_{1} & X_{2}& X_{3}& X_{4}& X_{5} \\
    X_{1} & 1 & r_{12} & r_{13} &r_{14} &r_{15} \\
    X_{2} &  & 1 & r_{23} & r_{24} & r_{25} \\
    X_{3} & & & 1 & r_{34} & r_{35} \\
    X_{4} & 대 & 칭 & & 1 & r_{45} \\
    X_{5} & & & & & 1
\end{matrix}$$

이 때 두 행렬 모두 상삼각과 하삼각의 원소가 일치하는 대칭(symmetric)행렬이 된다.
$$ s_{ij} = s_{ji}$$
이기 때문이다.
또한 대각 원소에는 분산 공분산 행렬에서는 자신과의 공분산 즉 분산이 오며 상관행렬에는 자신과의 상관계수 즉 1이 온다.

n x n 대칭행렬에는 고유값(고유근)과 고유벡터의 쌍이 n개 존재한다. <vr>
상관 행렬 S에도 마찬가지이다.
$$Sa_i = \lambda_{i}a_i$$
$$a_i = \begin{bmatrix} a_{i1} \\a_{i2} \\ .\\.\\.\\a_{in} \end{bmatrix},  i = 1, 2, ..., n$$ 
와 같은 식이 성립한다. <br><br>
이의 특성 방정식
$$det(S-\lambda I) = |S-\lambda I| = 0$$
을 통해 n쌍의 고유값과 고유벡터의 쌍을 구한다.
$$(\lambda_1, a_1), (\lambda_2, a_2), ..., (\lambda_n, a_n)$$

고유벡터들은 서로 정규직교하는(독립인) 성질을 만족한다. <br> <br>
**직교화 성질**
$$a^{T}_{i}a_i = a^{2}_{i1} + a^{2}_{i2} + ... + a^{2}_{in} = 1$$
$$a^{T}_{i}a_j = a_{i1}a_{j1} + a_{i2}a_{j2} + ... + a_{in}a_{jn} = 0$$
$$ i, j = 1, 2, ... , n, i \ne j$$


즉, 고유벡터들은 길이가 1이며 서로 다른 고유벡터의 내적은 0이 된다.

이에 따라 새로운 변수(주성분)은 다음과 같이 고유벡터 ai의 원소를 가중치로 하는 선형결합으로 나타내어진다.
$$ PC_1 = a_1^T\textbf{X} = a_{11}X_1 + a_{12}X_2 + ... + a_{1n}X{n}$$
$$ PC_1 = a_2^T\textbf{X}$$
$$ ...$$
$$ PC_n = a_n^T\textbf{X}$$
<br>
이 때 각 주성분의 분산이 바로 해당 고유벡터와 쌍을 이루는 고유값이다.
$$ Var(PC_n) = Var(a^T_nX)$$
분산의 성질에 의해
$$ Var(a^T_n\textbf{X}) = a^T_n\textbf{S}a_n$$
이 때 $Sa_i = \lambda_ia_i$이고 $a^T_na_n = 1$이므로
$$a^T_nSa_n = \lambda_na^T_na_n = \lambda_n$$


두 주성분 사이의 공분산은
$$Cov(PC_i,PC_j) = Cov(a^T_i\textbf{X}, a^T_j\textbf{X}) = a^T_i\textbf{S}a_j$$
이 때 $Sa_i = \lambda_ia_i$이고 $a^T_ia_j = 0$이므로
$$a^T_i\textbf{S}a_j = a^T_i\lambda_ja_j = 0, i \ne j$$

최종적으로 주성분의 분산은 고유값이며 주성분은 공분산이 0이므로 서로 독립임을 알 수 있다.

고유값의 합은 기존 n x n 행렬의 대각원소의 합(trace)과 같다. <br>
즉 여기서는 람다의 합이 R의 대각원소의 합과 같다는 것이다. <br>
즉, 주성분의 분산의 합$\sum{\lambda_i}$은 원래 변수의 분산의 합$\sum{s_{ii}}$과 일치한다.

$$\lambda_1 >= \lambda_2 >= ... >= \lambda_n$$

## 표본상관행렬을 이용한 주성분분석

앞에서 배운 건 분산, 공분산 행렬 S에 기반한 주성분분석이다.
원래 변수 X를 표준화시킨 변수 벡터 Z에 대한 분산과 공분산은
$$Var(Z_j) = 1$$
$$Cov(Z_j, Z_k) = r_{jk} = Corr(X_j, X_k), j \ne k = 1, 2, ... p$$
이 표준화된 변수 벡터에 기초한 R에 기초한 주성분은 S로부터의 주성분의 기본성질을 그대로 유지하지만 유도되는 주성분은 다르다. <br>
서로 다른 단위를 가진 원소들의 주성분의 크기에 의미를 부여하기 힘들기 때문에 의미를 갖기 위해 표준화해준다.

## 주성분의 수 결정 기준

$\sum{\lambda_{i}} = \sum{s_{ii}}$이다. <br>
즉 람다의 총합은 총 변동이라고 볼 수 있는데 따라서 람다를 람다의 총합으로 나눠주면 총변동에 대한 해당 PC의 비율을 알 수 있다.
$$\lambda_i \over \sum{\lambda_{i}}$$
이 비율을 통해 정보의 손실을 감수하고 비율이 낮은 주성분을 제거할 수 있다.

### 총변이에 대한 공헌도

일반적으로 주성분의 누적 정보가 전체의 80%를 넘는다면 나머지 주성분을 무시한다.

### 개별 고유값의 크기 - 표준화의 경우

모든 설명변수 X가 서로 독립이라면 모든 PC는 원래 변수의 분산 s와 같은 크기의 변동을 갖는다. <br>
만약 주성분이 상관행렬에 기초(모든 변수가 표준화)한다면 상관행렬의 대각원소가 1이므로 각 주성분의 분산은 1이 된다. <br>
대응하는 람다가 1보다 작은 주성분은 원래 변수보다 작은 변이를 가지므로 고려할 가치가 없다고 생각되어 제외한다.

### scree plot

스크리 플롯은 x축에 주성분 번호, y축에 해당 주성분에 대응하는 고유값을 연결한 플롯으로 이 플롯의 어느 구간에서 기울기가 가파르게 떨어진 후 완만하게 감소한다면 그 하강지점에 대응되는 개수의 주성분을 포함한다.

## 1992년까지 경제지표 주성분 분석

```r
economic = read.csv("econo.csv", header = T)
```

```r
str(economic)
```

```r
eco.data = economic[, -1]
rownames(eco.data) = economic$year
```

연도를 행이름으로 지정한다.

```r
round(cor(eco.data), 2)
```

```r
library(aplpack)
```

```r
stars(eco.data, key.loc = c(3, 12))
```

```r
faces(eco.data)
```

우리 나라 경제가 점점 성장하는 것을 볼 수 있다.

```r
eco.pca = princomp(eco.data, cor = T, scores = T)
```

princomp 함수를 통해 주성분 분석을 할 수 있다.
<br> cor 패러미터로 표준화 여부를 고를 수 있다.
<br> scores 패러미터는 각 주성분의 값을 계산할지 여부에 대한 옵션이다.

```r
summary(eco.pca)
```

standard deviation 즉 표준편차는 람다(주성분의 분산)의 제곱근이다. <br>
proportion of variance는 std의 제곱을 람다의 총합으로 나눈 즉, 주성분의 변동 비율이다. <br>
cumulative proportion은 proportion of variance의 누적비율이다. 

 총변이에 대한 비율을 기준으로 하거나 1보다 큰 값을 기준으로 해도 PC1 하나면 충분하다.

```r
names(eco.pca)
```

loadings에는 가중치 값(고유벡터)이 저장되어 있다.

```r
screeplot(eco.pca, type = "line", pch = 19, main = "Scree Plot")
```

```r
eco.pca$loadings
```
