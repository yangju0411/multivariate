# Factor Analysis

## 인자 분석이란

여러 개의 서로 관련 있는 변수들로 측정된 자료에서 그 변수들을 공통적으로 설명할 수 있는 인자를 구하는 것 <br>
변수들은 인자의 선형 결합으로 나타내어진다.

## PCA와의 차이

둘 다 새로운 변수가 생성된다는 공통점이 있지만 차이가 있다. <br>

1. 주성분은 변수들의 선형결합으로 나타내어지는 반면 인자 분석은 변수들이 인자의 선형결합으로 나타내어짐 <br>
2. 인자 분석에는 오차항이 존재 <br>
3. 인자의 순서는 주성분과 달리 의미가 없다. <br>
4. 인자 분석의 한 방법 중에는 PCA가 존재하며 구해진 인자는 주성분과 동일

## 모델

$$\textbf{X}=(X_1, X_2, ..., X_p)^T$$
$$Var(X_i)=\sigma_{ii}=l_{i1}^2+l_{i2}^2+...+l_{ip}^2+\varphi_i$$
여기서 $\textbf{$l_{i}^2$}$는 공통분산 $\textbf{$\varphi$}$는 특수분산이다.
$$Var(\textbf{X})=\sum_{p \times p}=\begin{bmatrix} & & & \\ & & \huge{\sigma_{ii}}& \\ & & & \\ & & & \end{bmatrix}$$
$$E(\textbf{X})=\textbf{$\mu$}=(\mu_1,\mu_2,...,\mu_p)^T$$

p개의 변수를 설명하는 q개의 인자가 있을 때 <br>
$$X_1-\mu_1=l_{11}*f_1+l_{12}*f_2+...+l_{1q}*f_q+\epsilon_1$$
$$X_2-\mu_2=l_{21}*f_1+l_{22}*f_2+...+l_{2q}*f_q+\epsilon_2$$
$$...$$
$$X_p-\mu_p=l_{p1}*f_1+l_{p2}*f_2+...+l_{pq}*f_q+\epsilon_p$$
<br>
행렬로 나타내면
$$\textbf{X}-\textbf{$\mu$}=\textbf{LF}+\textbf{$\epsilon$}$$
이 때 **L**은 인자의 부하값 행렬 Factor Loadings, F는 인자의 벡터 Common Factor, $\epsilon$은 오차항 즉 특수 인자의 벡터이다.

## 인자분석의 가정

$\textbf{X}~N(\mu, \sum)$ 변수는 정규분포를 따른다. <br>
$E(f_j)=0, j=1,2,...,q$, $E(\textbf{F})=\textbf{0}$ 인자의 평균은 0이다.<br>
$E(\epsilon_i)=0, i=1,2,...,p$, $E(\textbf{$\epsilon$})=\textbf{0}$ 특수 인자의 평균도 0이다. <br>
$Var(f_i)=1, Cov(f_i,f_j)=0$, $Cov(\textbf{F})=\textbf{I}$ 인자의 분산은 1, 인자 간의 공분산은 0이다. 즉 인자는 서로 독립이다. <br>
$Var(\epsilon_i)=\varphi_i, Cov(\epsilon_i,\epsilon_j)=0$, $Cov(\textbf{$\epsilon$})=\textbf{$\phi$}=\textbf{$\varphi^T$I}$ 특수인자도 서로 독립이며 분산은 $\varphi_i,i=1,2,...p$이다.
$Cov(\textbf{$\epsilon$},\textbf{F})=\textbf{0}$ 특수인자와 인자는 서로 독립이다.

values는 고유값을 나타내준다. 주성분 분석과 일치한다.

주성분분석과 마찬가지로 고유값이 1이상, 스크리플롯의 기울기로 인자의 수를 선택해준다.

```r
med.varimax
```

u2: 고유분산 $u2=1-h2$

h2 of lung: $0.47 = 0.66^2 + 0.12^2 + 0.16^2$

SS loadings of RC1: $2.39 = 0.66^2 + 0.11^2 + 0.78^2 + ... + (-0.07)^2$

F2-stamina, stretch, blow, urine)
F3-muscle, skeleton)
